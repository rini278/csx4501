```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import trim_mean
from statsmodels import robust

```

```python
mydata_new['voted_pres_16'].value_counts().sort_values().plot(kind='barh')
```

```python
print(mydata_new['age'].mean())
```

```python
print(trim_mean(mydata_new['age'],0.1))
```

```python
print(mydata_new['age'].median())
```

```python
print(mydata_new['age'].std())
```

```python
print(mydata_new['age'].quantile(0.75) - mydata_new['age'].quantile(0.25))
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
ax = sns.heatmap(mydata_new.corr(), vmin=-1, vmax=1, 
                 cmap=sns.diverging_palette(20, 220, as_cmap=True),
                 ax=ax)

plt.tight_layout()
plt.show()
```

```python
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(mydata_new, col ="employ",  hue ="voted_pres_16")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "voted_pres_16", "employ", edgecolor ="w").add_legend()
# show the object
plt.show()
  
# This code is contributed by Deepanshu Rustagi.
```
